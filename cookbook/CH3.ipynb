```python
%pylab inline 
import pandas as pd
import seaborn as sns
```

```python
bicycle = pd.read_csv('./../data/bikes.csv',sep=';',encoding='latin',index_col='Date',parse_dates='Date',
                     dayfirst=True)
```

```python
bicycle['weekday']=bicycle.index.dayofweek
```

```python
weekday=bicycle.groupby('weekday')['Berri 1'].aggregate(sum)
```

Note that the use of sum rather then count to sum the values for each columns rather than count to just count how man rows there are. 

```python
weekday.index = ['Mon','Tu','Wed','Thr','Fri','Sat','Sun']
```
